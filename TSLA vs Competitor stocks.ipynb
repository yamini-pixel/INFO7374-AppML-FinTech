```python
import numpy as np
from scipy.optimize import minimize
import scipy.stats as stats
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from datetime import datetime
import yfinance as yf
```

# **KDE** **plottings**

```python
start_date = datetime(2023,1,1) # 2019 1 1
end_date = datetime(2023,8,31)

plt.figure(1)
STOCK = yf.download('TSLA',start_date ,end_date)
sns.kdeplot(data=STOCK['Adj Close'].values, linewidth=2, color='grey')
print(STOCK['Adj Close'].describe())

# Calculate skewness, kurtosis and variance
print('Skewness:',stats.skew(STOCK["Adj Close"]))
print('Kurtosis:',stats.kurtosis(STOCK['Adj Close']))
print('Variance:',np.var(STOCK['Adj Close']))
```

```python
# Increasse the data sampling frequency
plt.figure(2)
MONTH = STOCK.resample('M').mean()
sns.kdeplot(data=MONTH['Adj Close'], linewidth=2, color='grey')
MONTH['Adj Close'].describe()
```

# **Visualize TSLA stock data from 1 Jan 2023 to 31 Aug 2023**

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Define the stock symbol and date range

start_date = datetime(2023,1,1)
end_date = datetime(2023,8,31)

# Fetch historical stock data from Yahoo Finance
tesla_stock_data = yf.download('TSLA', start_date, end_date)

# Create a basic stock price chart
plt.figure(figsize=(12, 6))
plt.title("TSLA Stock Price History from 1 Jan 2023 to 31 Aug 2023")
plt.xlabel("Date")
plt.ylabel("Price (USD)")

# Plotting different stock attributes

plt.plot(tesla_stock_data.index, tesla_stock_data["Open"], label="Open", color="blue")

plt.plot(tesla_stock_data.index, tesla_stock_data["Close"], label="Close", color="green")

plt.plot(tesla_stock_data.index, tesla_stock_data["High"], label="High", color="red")

plt.plot(tesla_stock_data.index, tesla_stock_data["Low"], label="Low", color="orange")

plt.legend()
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```
